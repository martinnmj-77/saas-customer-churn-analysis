```python
import pandas as pd
df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.isnull().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
df= df.dropna()
```

```python
df['Churn'].value_counts(normalize=True)
```

```python
#Churn rate overall
df['Churn'].value_counts(normalize=True)*100
```

```python
#Churn by Tenure
df.groupby('tenure')['Churn'].value_counts(normalize=True).unstack()
```

```python
df['tenure_group'] = pd.cut(df['tenure'], bins = [0,12,24,48.72])
df.groupby('tenure_group')['Churn'].value_counts(normalize=True).unstack()
```

```python
#Churn by Monthly Charges
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
#Churn by Contract Type
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack()
```

```python
#Churn by Payment Method
df.groupby('PaymentMethod')['Churn'].value_counts(normalize=True).unstack()
```
